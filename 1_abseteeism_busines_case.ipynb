"Reason for Absence" is actually a categorical variable as each number represents some disease.
Since each person had one reason we can convert these 0-28 numbers into 0 and 1, that is into dummies:
a given reason appeared (1) or not (0).

```python
reason_columns = pd.get_dummies(df["Reason for Absence"])
reason_columns
```

```python
# Checking for missing values
reason_columns["check"] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns["check"].sum(axis=0)
```

```python
reason_columns["check"].unique()
```

```python
# If everything is OK, we can remove "check" column
reason_columns = reason_columns.drop(["check"], axis=1)
```

```python
# To avoid potential multicollinearity in the model we have to remove column "0". This stands for unknown reason
# and can be set as a baseline of the analysis.
reason_columns = pd.get_dummies(df["Reason for Absence"], drop_first=True)
reason_columns
```

```python
# Because all reasons are now stored in separate columns, we have to remove "Reason for Absence" from the
# original dataframe (this is to avoid multicollinearity)
df = df.drop(["Reason for Absence"], axis=1)
```

```python
# There are many dummies (27), it is much efficient to group (or classify) them. The grouping is based on 
# a description of the diseases
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:28].max(axis=1)
```

```python
# Adding the new column to df:
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
```

```python
# We want to rename the added columns
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
                'Daily Work Load Average', 'Body Mass Index', 'Education',
                'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 
                'Reason_2', 'Reason_3', 'Reason_4'
               ]
```

```python
df.columns = column_names
df
```

```python
# Reorder columns
reordered_column_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                          'Date', 'Transportation Expense', 'Distance to Work', 
                          'Age', 'Daily Work Load Average', 'Body Mass Index', 
                          'Education', 'Children', 'Pets', 'Absenteeism Time in Hours'
                         ]
```

```python
df = df[reordered_column_names]
df.head()
```

## Creating a checkpoint

```python
# This is important to save the current state of the code. It may be needed if our next changes appear not 
# correct and we want to get back to the state which works
df_reason_mod = df.copy()
```

## Date

```python
df_reason_mod["Date"]
```

```python
type(df_reason_mod["Date"][0])
```

```python
# Converting to datetime format
df_reason_mod["Date"] = pd.to_datetime(df_reason_mod["Date"], format="%d/%m/%Y")
df_reason_mod["Date"]
```

```python
# Extracting day and month as separate columns
list_months = []
list_days = []

for i in range(len(df_reason_mod["Date"])):
    list_months.append(df_reason_mod["Date"][i].month)
    list_days.append(df_reason_mod["Date"][i].day)
    
list_months
len(list_months)
len(list_days)
```

```python
df_reason_mod["Month value"] = list_months
```

```python
# Extracting day of the week
df_reason_mod["Date"][699].weekday()

# In this module weekdays are from 0 to 6
```

```python
# We create a function which can be applied to each row to extract weekday
def date_to_weekday(date_value):
    return date_value.weekday()

df_reason_mod["Day of the week"] = df_reason_mod["Date"].apply(date_to_weekday)
df_reason_mod.head()
```

```python
# We drop 'Date' column and reorder the frame
df_reason_mod = df_reason_mod.drop(["Date"], axis=1)
```

```python
df_reason_mod.columns.values
```

```python
reordered_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                   'Month value', 'Day of the week', 'Transportation Expense', 
                   'Distance to Work', 'Age', 'Daily Work Load Average', 
                   'Body Mass Index', 'Education', 'Children', 'Pets', 
                   'Absenteeism Time in Hours'
                  ]
```

```python
df_reason_mod = df_reason_mod[reordered_names]
```

```python
df_reason_mod.head()
```

## Creating next checkpoint

```python
df_reason_date_mod = df_reason_mod.copy()

# We do not need to modify 'Transportation Expenses', 'Distance to Work', 'Age', 'Daily Work Load Average', and 
# 'Body Mass Index'.
```
